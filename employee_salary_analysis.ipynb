```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Employee": ["John", "Bob", "Alice", "John", "David", "Emma"],
    "Departments": ["IT", "IT", "HR", "IT", "HR", "Finance"],
    "Salary": [50000, 60000, 45000, 50000, 48000, 55000]
})

print(df)

df = df.drop_duplicates().reset_index(drop = True)
print("Average Salary = ", df["Salary"].mean())
print("Maximum Salary = ", df["Salary"].max())
print("Minumum Salary = ", df["Salary"].min())
print(df.groupby("Departments")["Salary"].agg(["mean", "count", "sum"]))
print(pd.pivot_table(df, values = "Salary", index = "Departments", columns ="Employee"))
print(df.sort_values("Salary", ascending=False))
df.groupby("Departments")["Salary"].sum().plot(kind="bar")
plt.title("Total Salary by Department")
plt.xlabel("Department")
plt.ylabel("Total Salary")
plt.show()
df.to_csv("employee_cleanup_data.csv", index = False)

```
